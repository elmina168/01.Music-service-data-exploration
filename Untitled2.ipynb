# **География успеха: что делает трек виральным в разных странах?**

Топ-жанры по виральности в разрезе стран (только rising).

```python
rising_only = df[df['trend'] == 'Rising']
top_genre_by_country = rising_only.groupby('country').apply(
    lambda x: x.groupby('genre')['viral_score'].mean().idxmax()
).reset_index(name='top_rising_genre')
```

```python
avg_score = rising_only.groupby(['country', 'genre'])['viral_score'].mean().reset_index()
top_genre_by_country = top_genre_by_country.merge(
    avg_score,
    left_on=['country', 'top_rising_genre'],
    right_on=['country', 'genre']
).rename(columns={'viral_score': 'avg_viral_score'}).drop('genre', axis=1)
```

```python
print('\n Топ-жанры по вриальности среди растущих треков (Rising)')
print(top_genre_by_country.to_string(index=False))
```

Доля rising/falling по странам.

```python
trend_by_country = df.groupby(['country', 'trend']).size().unstack(fill_value=0)
trend_by_country['total'] = trend_by_country.sum(axis=1)
for trend_col in ['Rising', 'Falling']:
    if trend_col in trend_by_country.columns:
        trend_by_country[f'% {trend_col}'] = (trend_by_country[trend_col] / trend_by_country['total']) * 100
    else:
        trend_by_country[f'% {trend_col}'] = 0

trend_by_country = trend_by_country.sort_values('% Rising', ascending=False)
```

```python
print('\nДоля Rising/Falling по странам')
print(trend_by_country[['% Rising', '% Falling']].round(1).head(10))
```

Корреляция viral_score с другими метриками по странам.




```python
countries = df['country'].unique()
corr_results = []

for country in countries:
    subset = df[df['country'] == country]
    if len(subset) > 5:  # минимум 5 песен
        corr_streams, _ = pearsonr(subset['viral_score'], subset['streams'])  # исправлено: было stream_change
        corr_pos, _ = pearsonr(subset['viral_score'], subset['pos'])
        corr_days, _ = pearsonr(subset['viral_score'], subset['days'])
        corr_change, _ = pearsonr(subset['viral_score'], subset['stream_change'])

        corr_results.append({
            'country': country,
            'corr_streams': corr_streams,
            'corr_pos': corr_pos,
            'corr_days': corr_days,
            'corr_stream_change': corr_change,
            'n_tracks': len(subset)
        })

corr_df = pd.DataFrame(corr_results)
```

```python
print('\nКорреляция viral_score с другими метриками по странам.')
print(corr_df.round(2).to_string(index=False))
```

Визуалиация.

```python
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

#1
plt.figure(figsize=(12, 6))
top_countries = trend_by_country.head(10).reset_index()
sns.barplot(data=top_countries, x='country', y='% Rising', palette='viridis')
plt.title('Доля растущих песен по странам (Rising)', fontsize=14)
plt.xlabel('Страна')
plt.ylabel('% Rising песен')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show
```

```python
#2
top5_countries = corr_df.nlargest(5, 'n_tracks')['country'].values
corr_matrix_top5 = corr_df[corr_df['country'].isin(top5_countries)].set_index('country')
corr_matrix_top5 = corr_matrix_top5[['corr_streams', 'corr_pos', 'corr_days', 'corr_stream_change']]

plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix_top5, annot=True, cmap='coolwarm', center=0,
            fmt='.2f', linewidths=0.5, cbar_kws={'label': 'Корреляция Пирсона'})
plt.title('Корреляция viral_score с метриками (топ-5 стран по числу треков)', fontsize=14)
plt.ylabel('Страна')
plt.xlabel('Метрика')
plt.tight_layout()
plt.show()
```

```python
#3
for target_country in ['US', 'MX']:
  subset = df[df['country'] == target_country]
  genre_viral = subset.groupby('genre')['viral_score'].mean().sort_values(ascending=False).head(8)

  plt.figure(figsize=(10, 5))
  genre_viral.plot(kind='bar', color='skyblue')
  plt.title(f'Средний viral_score по жанрам в {target_country}', fontsize=14)
  plt.xlabel('Жанр')
  plt.ylabel('Средний viral_score')
  plt.tight_layout()
  plt.show()
```

```python
#4
us_data = df[df['country'] == 'US'].copy()
plt.figure(figsize=(10, 6))
sns.scatterplot(data=us_data, x='streams', y='viral_score', hue='trend',
                size='days', sizes=(20, 200), alpha=0.7, palette={'Rising': 'green', 'Falling': 'red'})
plt.xlabel('Потоки (streams)')
plt.ylabel('Виральный скор (viral_score)')
plt.legend(title='Тренд', bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show
```
